```python
import numpy as np
# Update the values in the contributing factors columns to be either True, False, or NaN
def update_vals(x):
    if x == '-':
        return False
    elif pd.isnull(x):
        return np.nan
    else:
        return True
```

```python
tafe_dissat_cols = ['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']
```

```python
tafe_resignations['dissatisfied'] = tafe_resignations[tafe_dissat_cols].applymap(update_vals).any(axis=1, skipna=False)
```

```python
tafe_resignations['dissatisfied'].value_counts(dropna=False)
```

```python
tafe_resignations_up = tafe_resignations.copy()
```

# added a column 'dissatisfied' indicating that the employee hated his job and quit because it sucked

```python
dete_resignations_up['institute'] = 'DETE'
```

```python
tafe_resignations_up['institute'] = 'TAFE'
```

```python
# Combine the dataframes
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)
```

```python
combined.shape
```

```python
# Verify the number of non null values in each column
combined.notnull().sum().sort_values()
```

```python
#drop columns with less than 500 non-null values
combined_updated = combined.dropna(axis=1, thresh=500)
```

```python
combined_updated
```

# clean the service column

```python
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype(str).str.extract(r'(\d+)')
combined_updated
```

```python
#turn them into floats
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')
```

```python
def map_year_to_career_stage(year):
    if pd.isnull(year):
        return np.nan
    elif year < 3:
        return 'New'
    elif year <= 6:
        return 'Experienced'
    elif year <= 10:
        return 'Established'
    else:
        return 'Veteran'
```

# add service categories based on years with company

```python
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(map_year_to_career_stage)
```

```python
combined_updated
```

```python
combined_updated['dissatisfied'].value_counts(dropna=False)
```

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
```

```python
combined_updated['dissatisfied'].value_counts(dropna=False)
```

# Since a True value is considered to be 1 in a pivot table, calculating the mean will also calculate the percentage of dissatisfied employees.

```python
dissat_pct = combined_updated.pivot_table(index='service_cat', values='dissatisfied')
```

```python
# Plot the results
%matplotlib inline
dissat_pct.plot(kind='bar', rot=30)
```

From the initial analysis above, we can tentatively conclude that employees with 7 or more years of service are more likely to resign due to some kind of dissatisfaction with the job than employees with less than 7 years of service.
